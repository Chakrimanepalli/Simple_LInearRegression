# Delivery_time

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
# Load the dataset
file_path = "C:\\Users\\chakri\\Downloads\\delivery_time.csv"
df = pd.read_csv(file_path)
```

```python
# Visualize the distribution of Delivery Time using a box plot
sns.boxplot(x=df['Delivery Time'])
plt.title('Box Plot for Delivery Time')
plt.show()
```

```python
# Visualize the distribution of Sorting Time using a box plot
sns.boxplot(x=df['Sorting Time'])
plt.title('Box Plot for Sorting Time')
plt.show()
```

```python
# Scatter plot between Delivery Time and Sorting Time
plt.scatter(x=df['Delivery Time'], y=df['Sorting Time'])
plt.xlabel("Delivery Time")
plt.ylabel("Sorting Time")
plt.title('Scatter Plot: Delivery Time vs Sorting Time')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()
```

```python
# Display the correlation matrix
print(correlation_matrix)
```

```python
# Plot the distribution of Delivery Time
sns.displot(x=df['Delivery Time'], kde=True)
plt.title('Distribution of Delivery Time')
plt.show()
```

```python
# Fit a linear regression model
model = smf.ols("Q('Delivery Time') ~ Q('Sorting Time')", data=df).fit()
```

```python
# Display the regression summary
print(model.summary())
```

```python
# Plot a regression line
sns.regplot(x=df['Delivery Time'], y=df['Sorting Time'], data=df)
plt.title('Regression Plot: Delivery Time vs Sorting Time')
plt.show()
```

```python
# Predict values using the fitted model
predict_values = model.predict()
```

```python
# Display the predicted values
print(predict_values)
```

# Salary_hike

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
# Load the dataset
file_path = "C:\\Users\\chakri\\Downloads\\Salary_Data.csv"
df = pd.read_csv(file_path)
```

```python
# Visualize the distribution of Salary using a box plot
sns.boxplot(x=df['Salary'])
plt.title('Box Plot for Salary')
plt.show()
```

```python
# Visualize the distribution of YearsExperience using a box plot
sns.boxplot(x=df['YearsExperience'])
plt.title('Box Plot for YearsExperience')
plt.show()
```

```python
# Scatter plot between Salary and YearsExperience
plt.scatter(x=df['Salary'], y=df['YearsExperience'])
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title('Scatter Plot: Salary vs YearsExperience')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()
```

```python
# Display the correlation matrix
print(correlation_matrix)
```

```python
# Plot the distribution of Salary
sns.displot(x=df['Salary'], kde=True)
plt.title('Distribution of Salary')
plt.show()
```

```python
# Fit a linear regression model
model = smf.ols("Q('YearsExperience') ~ Q('Salary')", data=df).fit()
```

```python
# Display the regression summary
print(model.summary())
```

```python
# Plot a regression line
sns.regplot(x=df['Salary'], y=df['YearsExperience'], data=df)
plt.title('Regression Plot: Salary vs YearsExperience')
plt.show()
```

```python
# Predict values using the fitted model
predict_values = model.predict()
```

```python
# Display the predicted values
print(predict_values)
```
